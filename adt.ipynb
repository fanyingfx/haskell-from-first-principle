```haskell
data OperatingSystem =  GnuPlusLinux  | OpenBSDPlusNevermindJustBSDStill  | Mac  | Windows  deriving (Eq, Show)  
data ProgrammingLanguage =  Haskell  | Agda  | Idris  | PureScript  deriving (Eq, Show)  
data Programmer =  Programmer { os :: OperatingSystem  , lang :: ProgrammingLanguage }  deriving (Eq, Show)
```

```haskell
allOperatingSystems :: [OperatingSystem]  
allOperatingSystems =  [ GnuPlusLinux  , OpenBSDPlusNevermindJustBSDStill  , Mac  , Windows  ]  
allLanguages :: [ProgrammingLanguage]  
allLanguages = [Haskell, Agda, Idris, PureScript]

allProgrammers = [Programmer x y | x<-allOperatingSystems , y<-allLanguages]
-- length allProgrammers
gen_all :: [OperatingSystem]->[ProgrammingLanguage]->[Programmer]
gen_all x y = go x y []
    where 
        go [] _ res  = res
        go (o:os) ys res = go os ys ((go' o ys []) ++ res) where
                    go' _ [] res = res
                    go' o (y:ys) res = go' o ys ((Programmer o y):res)

length (gen_all allOperatingSystems allLanguages)
```

## BinaryTree

```haskell
data BinaryTree a= 
    Leaf
    | Node (BinaryTree a) a ( BinaryTree a)
    deriving (Eq,Ord,Show)
```

```haskell
insert' :: Ord a => a -> BinaryTree a -> BinaryTree a
insert' b Leaf = Node Leaf b Leaf
insert' b (Node left a right)
    | b == a = Node left a right
    | b < a = Node (insert' b left) a right
    | b > a = Node left a (insert' b right)
```

```haskell
t1 = insert' 0 Leaf
t2 = insert' 3 t1
t3 = insert' 5 t2
t3
```

```haskell
mapBinaryTree :: (a->b)->BinaryTree a -> BinaryTree b
mapBinaryTree _ Leaf = Leaf
mapBinaryTree f (Node ltree a rtree) = Node (mapBinaryTree f ltree) (f a) (mapBinaryTree f rtree)
mapBinaryTree (+1) t3
```

```haskell
testTree' :: BinaryTree Integer  
testTree' =  Node (Node Leaf 3 Leaf) 1 (Node Leaf 4 Leaf)  
mapExpected =  Node (Node Leaf 4 Leaf) 2 (Node Leaf 5 Leaf)  
-- acceptance test for mapTree  
mapOkay =  if mapBinaryTree (+1) testTree' == mapExpected  then print "yup okay!"  else error "test failed!"
mapOkay
```

```haskell
foldTree :: (a->b->b) -> b -> BinaryTree a -> b
foldTree f b Leaf = b
foldTree f b (Node left a right) = foldTree f (foldTree  f  (f a b) left) right

foldTree (+) 0 t3
```

```haskell
import Data.Char

caesarChar :: Char -> Int -> Char
caesarChar c n
  | isAsciiLower c = chr (ord 'a' + (ord c - ord 'a' + n) `mod` 26)
  | isAsciiUpper c = chr (ord 'A' + (ord c - ord 'A' + n) `mod` 26)
  | otherwise = c

unCaesar :: Char -> Int -> Char
-- chr (ord 'a' + (ord c - ord 'a' - n) `mod` 26)
unCaesar c n
  | isAsciiLower c = chr (ord 'a' + (ord c - ord 'a' - n) `mod` 26)
  | isAsciiUpper c = chr (ord 'A' + (ord c - ord 'A' - n) `mod` 26)

offset :: Char -> Int
offset c = (ord c) - (ord 'A')

encrypt :: String -> String -> String
encrypt msg key = go msg 0 []
  where
    go [] _ res = res
    go (x : xs) n res = go xs (n + 1) (res ++ [caesarChar x (offset (key !! (n `mod` length key)))])
```

```haskell
encrypt "MEET AT DAWN" "ALLY"
```

```haskell
encryptOffset :: String -> String -> [(Char,Int)]
encryptOffset msg key = go msg 0 []
  where
    go [] _ res = res
    go (x : xs) n res = go xs (n + 1) ( res ++ [(x ,(offset (key !! (n `mod` length key))))])
encryptOffset "MEET AT DAWN" "ALLY"
```

```haskell
capitalizeParagraph :: String -> String
capitalizeParagraph = unwords . capitalize . words
    where
        capitalize [] = []
        capitalize wws@(w:ws) = capitalizeWord w : capitalizeAfterPeriod wws
        endsInPeriod [] = False
        endsInPeriod w = last w == '.'
        capitalizeAfterPeriod (wb:w:ws) =
            if endsInPeriod wb
                then capitalizeWord w : capitalizeAfterPeriod (w:ws)
                else w : capitalizeAfterPeriod (w:ws)
        capitalizeAfterPeriod _ = []
capitalizeParagraph "woot ha. cc."
```

```haskell
import Data.Char
import Data.List (groupBy)
type Digit=Char
type Button = (Digit, String)
buttons :: [Button]
buttons = [('1', "1"),
   ('2', "abc2"),
   ('3', "def3"),
   ('4', "ghi4"),
   ('5', "jkl5"),
   ('6', "mno6"),
   ('7', "pqrs7"),
   ('8', "tuv8"),
   ('9', "wxyz9"),
   ('*', "*^"),
   ('0', " +_0"),
   ('#', "#.,")]
data DaPhone = DaPhone [Button] 
phone = DaPhone buttons

-- validButtons = "1234567890*#"
-- Valid presses: 1 and up
type Presses=Int
reverseTaps  :: DaPhone -> Char -> [(Digit,Presses)]
reverseTaps (DaPhone buttons) c = if isUpper c then go (toLower c) buttons  [('*',1)] else go c buttons []
    where 
        findPress c "" n = n
        findPress c (x:xs) n= if x == c then n+1 else findPress c xs (n+1)
        go c []  res = res
        go c ((digit,seq):otherButtons)  res = if  c `elem` seq then res ++ [(digit, findPress c seq 0)] else go c otherButtons res
-- reverseTaps phone '1'
cellPhonesDead :: DaPhone -> String -> [(Digit,Presses)]
cellPhonesDead phone str  = foldr f  [] str
    where 
        f c acc = (reverseTaps phone c) ++ acc
-- cellPhonesDead phone "abc"

fingerTaps :: [(Digit,Presses)] -> Presses
fingerTaps xs  = sum $ (map snd xs)

groupByDigit :: [(Digit,Presses)] -> [[(Digit,Presses)]]
groupByDigit = groupBy (\x y -> (fst x) == (fst y))
groupByDigitAndSum :: [(Digit,Presses)]->(Digit,Int)
groupByDigitAndSum taps = (fst . head  $ taps , fingerTaps taps)
tapsCount  msg= map groupByDigitAndSum (groupByDigit (cellPhonesDead phone msg))
mostPopularDigit :: String -> Char
mostPopularDigit str = fst (foldr f (head str,0) (tapsCount str))
    where f x@(_,x_cnt) maxTap@(_,max_cnt) = if x_cnt > max_cnt then x else maxTap
mostPopularDigit "Hello"


```
